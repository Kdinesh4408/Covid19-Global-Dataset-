# DATA VISULIZATION :

# 1.Which highest continent spread COVID-19

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.countplot(x='continent',data=ad,ax=ax,color='darkred')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Highest Continent')
```

# 2.Comparison between confirmed,deaths,total_recovered and active cases with continent

```python
ad2=ad[['confirmed','deaths','total_recovered','active cases','continent']]
ad2
```

```python
sns.pairplot(ad2,hue='continent',height=2.5)
```

# 3.Analysis continent according total cases per 1m population

```python
plt.subplots(figsize=(10,6))
sns.boxplot(x=ad['continent'],y=ad['total cases per 1m population'],palette='rainbow')
```

# 4.continent according total deat per 1m population

```python
plt.subplots(figsize=(14,6))
sns.barplot(x=ad['total deat hs per 1m population'],y=ad['continent'],color='r')
plt.show()
```

```python
ad.head()
```

# 5.Comparison between total deat hs per 1m population and total cases per 1m population with country

```python
ad3 = ad[['total deat hs per 1m population','total cases per 1m population','country']]
ad3.head()
```

```python
sns.pairplot(ad3,hue='country',height=2.5)
```

# 6.Analysis Country according total tests

```python
plt.subplots(figsize=(10,20))
sns.barplot(x='total tests',y='country',data=ad,palette='rainbow')
```

# 7.Comparison between population and total tests

```python
plt.subplots(figsize=(25,6))
ax=sns.swarmplot(x=ad['population'],y=ad['total tests'],color='brown')
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

# 8.Which continent highest cases in percentage

```python
ad.groupby('continent')['country'].count().nlargest(8).plot(kind = 'pie', autopct = '%1.1f%%')
```
